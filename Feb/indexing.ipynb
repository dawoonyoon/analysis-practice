1. row number _ iloc

```python
import pandas as pd

df = pd.read_csv('data/uk-500.csv')
df = df.set_index('last_name')
df
```

```python
# row 기준
df.iloc[0]
```

```python
# column 기준
df.iloc[:, 0]
```

2. label, condition, index _ .loc

```python
# selecting row with Andrade _ return Series
df.loc['Andrade']
```

```python
# selecting Andrade & Veness
df.loc[['Andrade', 'Veness']]
```

```python
# selecting row and column 
df.loc[['Andrade', 'Veness'], ['first_name', 'address', 'city']]
```

```python
df.loc[['Andrade', 'Veness'], 'city':'email']
```

```python
df[df['first_name'] == 'Antonio']
```

```python
df.loc[df['first_name'] == 'Antonio', 'email']
```

```python
df.loc[df['first_name'] == 'Antonio', ['email']]
```

```python
import numpy as np
import pandas as pd
```

```python

# data를 dict 형태로 정의

data = {'Name':['s1', 's2', 's3'],
        'Age':[25, 28, 22],
        'Score':np.array([95, 80, 75])}
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'])
df
```

```python
# data를 list 형태로 정의

data2 = [['s1', 25, 95],
         ['s2', 28, 80],
         ['s3', 22, 75]]
df2 = pd.DataFrame(data2, 
                   index=['row1', 'row2', 'row3'],
                   columns=['Name', 'Age', 'Score'])
df2
```
