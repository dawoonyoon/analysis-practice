### 필요한 column만 추출해서 정리

```python
a1 = ['경기일자', '원정팀', '원정팀점수', '홈팀', '홈팀점수']
games_nae = df1_na[a1]
games_nae
```

```python
# 경기일자 column을 index로 지정
# append=True -> 이전에 있던 index도 유지
games_naes = games_nae.set_index('경기일자', append=True)
games_naes
```

```python
# index 이름 설정
games_naesr = games_naes.rename_axis(['게임', '경기일자'])
games_naesr
```

### melt 함수 사용해보기

```python
df = pd.DataFrame({
    'A' : ['a1', 'a2', 'a3', 'a4'],
    'B' : ['b1', 'b2', 'b3', 'b4'],
    'C' : [1,2,3,4],
    'D' : [100,200,300,400]
}) 
df
```

```python
df.melt(id_vars='A', value_vars=['B', 'C'])
```

```python
# index 다시 설정
a1 = games_naesr.reset_index()
a1
```

```python
tidy = pd.melt(a1, id_vars=['게임', '경기일자'], 
               value_vars=['원정팀', '홈팀'],
              var_name='팀구분', value_name='팀')
tidy.query('게임 == 0')
```

```python
# diff 함수
# 현재 값이 이전보다 얼마나 등락했는지를 계산해서 반환
a1 = tidy.query('팀 == "Boston Celtics"')
# 경기일자를 기준으로 오름차순 정리
a2 = a1.sort_values('경기일자')
# 날짜 간격
a2['경기일자'].diff()
```

```python
# 각 팀별로 그룹을 묶고, 그 안에서 전날보다 며칠 후에 경기를 했는지 계산
tidy.sort_values('경기일자', inplace=True)
a1 = tidy.groupby('팀')['경기일자'].diff()
# 날짜를 정수로
a2 = a1.dt.days
# 날짜 차이값이 1일 (연속으로 경기가 있었던 날)
tidy[['경기일자', '팀']][a2 == 1]
```

### 각 팀의 휴식일

```python
# 휴식일 확인 
ab = tidy.groupby('팀')['경기일자'].diff()
# 정수형으로 
ab = ab.dt.days
ab = ab - 1
ab
```

```python
# 결측치 확인
ab.isna().sum()
```

```python
# 결측치를 0으로 대체
ab.fillna(0,inplace=True)
ab
```

```python
tidy['휴식일'] = ab
# index를 기준으로 정렬
tidy.sort_index(inplace=True)
tidy
```
