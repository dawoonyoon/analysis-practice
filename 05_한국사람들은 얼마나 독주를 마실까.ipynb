```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 불필요한 경고 메세지를 나오지 않도록
import warnings
warnings.filterwarnings('ignore')

# Seaborn에서 제공하는 그래프 테마
# sns.set_style("ticks")
# sns.set_style("darkgrid")
# sns.set_style("whitegrid")

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

```python
drink = pd.read_csv('data/drinks.csv')
drink
```

```python
# 결측치 확인
drink.isna().sum()
```

```python
# NA in continent columns actually refers to North America
# 결측치를 NA로 설정
drink['continent'].fillna('NA', inplace=True)
drink.isna().sum()
```

```python
drink['continent'].value_counts()
```

```python
# 전체 술 소비량 계산
total_servings = drink['beer_servings'] + drink['spirit_servings'] + drink['wine_servings']
drink['total_servings'] = total_servings
drink
```

```python
# 술 소비량 대비 알콜비율 계산
a1 = drink['total_litres_of_pure_alcohol'] / drink['total_servings']
a1
```

```python
# 전체 술 소비량이 0인 국가는 계산이 불가능하므로 결과가 결측치로 나타난다
# 결측치를 0으로 대체 
a1.fillna(0, inplace=True)
a1.isna().sum()
```

```python
# 알콜비율을 DataFrame에 추가 
drink['alcohol_rate'] = a1
drink
```

```python
# 국가와 알콜비율을 추출
country_with_rank = drink[['country', 'alcohol_rate']]
country_with_rank
```

```python
# alcohol_rate를 기준으로 내림차순 정렬
country_with_rank.sort_values(by='alcohol_rate', ascending=False, inplace=True)
country_with_rank
```

```python
# x 좌표
x_pos = list(range(len(country_with_rank)))
print(x_pos)
```

```python
# y 좌표
rank = country_with_rank['alcohol_rate']
rank
```

```python
# 시각화
# 막대 그래프를 그리는 함수는 각 막대들을 제어할 수 있게 객체가 담긴 리스트를 반환한다
bar_list = plt.bar(x_pos, rank)
# print(bar_list)

# 한국에 해당하는 막대의 index를 가져오고, 국가 이름들을 추출
names = country_with_rank['country'].tolist()

# 한국의 index
korean_idx = names.index('South Korea')
# print(korean_idx)

# 한국에 해당하는 막대를 붉은색으로 설정
bar_list[korean_idx].set_color('r')

# annotation 화살표 설정
arrow_dict = {
    'facecolor' : 'red'
}

# 표시할 문자열, annotation의 위치
plt.annotate('South Korea', xy=(korean_idx, rank.values[korean_idx]),
            # annotation내의 화살표 옵션 설정
             arrowprops=arrow_dict,
             # annotation내의 문자열 위치
            xytext=(korean_idx + 10, rank.values[korean_idx] + 0.05))
plt.show()
```
